```python
import pandas as pd
import statsmodels.formula.api as smf
import statsmodels.api as sm
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('Data2.csv', sep=r'\s+')
data
```

## Rate

```python
# Rate
model_lm = smf.ols('Rate ~ Game + ThreeD + Game*ThreeD', data = data).fit()
```

```python
import scipy.stats as stats
plt.hist(model_lm.resid, edgecolor='black')
plt.show()
```

```python
g1 = data['Rate'][0:8]
g2 = data['Rate'][8:16]
g3 = data['Rate'][16:24]
g4 = data['Rate'][24:32]

residual = []
for r in g1:
    residual.append(r-g1.mean())
for r in g2:
    residual.append(r-g2.mean())
for r in g3:
    residual.append(r-g3.mean())
for r in g4:
    residual.append(r-g4.mean())
    
    
plt.hist(residual, edgecolor='black')
plt.show()
```

```python
# In Shapiro Wilk test p-value should be higher than 0.05 to show normality
stats.shapiro(residual)
```

```python
# In Shapiro Wilk test p-value should be higher than 0.05 to show normality
stats.shapiro(model_lm.resid)
```
